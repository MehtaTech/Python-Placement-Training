```python
def calculate_bill(hours):
    bill = 0
    
    if hours <= 2:
        bill = hours * 100
    elif hours <= 5:
        bill = (2 * 100) + (hours - 2) * 50
    else:
        bill = (2 * 100) + (3 * 50) + (hours - 5) * 25
    
    return bill

# Example usage
h = int(input("Enter parking hours: "))
print("Total Bill = ₹", calculate_bill(h))
```

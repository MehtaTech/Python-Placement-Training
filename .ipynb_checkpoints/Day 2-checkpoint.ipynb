### You own a praking plot, you have different prcing for different number of hours, You need to hire a devloper  to calculate the total bill with the help of a program
#### 100 for first 2 hour 100 rupee per hour
#### For next 3 hours 50 rupees per hour
#### Rest or remaining hours 25 rupees per hour

```python
hours = float(input("Enter number of hours the vehicle was parked :- "))
bill = 0

if hours:
        if hours <= 2:  
            bill = hours * 100
            print("Bill = ",bill)
        elif hours <= 5:
            bill =(2 * 100) + (hours - 2) * 50
            print("Bill = ",bill)
        else:
            bill = (2*100) + 3 * 50 +(hours - 5) * 25
            print("Bill = ",bill)



    
```

## You need to calculate the total bill based on number of units
#### 0 - 100 units :- 1.5 per unit
#### 101-200 units :- 3.5 per unit
#### Remaining units 5.0 per unit, you need to add rupees 50 as a fixed charge, if bill exceeds rupees 2000 then add a surplus charge of 10% of total bill

```python
units = int(input("Enter number of units: "))
bill = 0


if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = (100 * 1.5) + (units - 100) * 3.5
else:
    bill = (100 * 1.5) + (100 * 3.5) + (units - 200) * 5.0

bill += 50

if bill > 2000:
    bill += bill * 0.10

print("Total Bill = ", bill)
```
